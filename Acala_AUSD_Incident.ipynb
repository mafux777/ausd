Let's start with some utilities for substrate for python..

```python
# next line only needed in Google Colaboratory! Otherwise pip install yourself
!pip install substrate-interface

from substrateinterface import SubstrateInterface

interlay_substrate = SubstrateInterface(
    url="wss://api.interlay.io/parachain"
)
dot_substrate = SubstrateInterface(
    url="wss://polkadot-rpc.dwellir.com"
)

def convert_intr_to_dot(intr):
    universal = interlay_substrate.ss58_decode(intr)
    polkadot = dot_substrate.ss58_encode(universal)
    return polkadot

def convert_dot_to_interlay(dot):
    universal = dot_substrate.ss58_decode(dot)
    dot = interlay_substrate.ss58_encode(universal)
    return dot

```

reward contract?)

```python
#address = "26JmEcghNmggvT46sojckg34Py9zFRKkCcFy3gr49hrFgT2k"
address = "23M5ttkmR6KcnxentoqchgBdUjzMDSzFoUyf5qMs7FsmRMvV"

url = f"https://api.polkaholic.io"
next_page = f"/account/transfers/{address}"

my_data = []
for _ in range(10):
  r = requests.get(url=url+next_page, headers=headers).json()
  my_data.extend(r['data'])
  next_page = r.get('nextPage')
  print(next_page)
  if next_page:
    pass
  else:
    break


xcm = pd.json_normalize(my_data).sort_values('ts', ascending=False)
xcm['timestamp'] = pd.to_datetime(xcm.ts, unit="s")
interesting = ['timestamp',*xcm.columns]
```

```python
stats = xcm.loc[xcm.rawAsset=='{"token":"AUSD"}'].groupby(['chainID', 'id', 'transferType','from', 'to', 'rawAsset']).agg({
    'amountUSD' : 'sum',
    'timestamp' : 'max'
})
stats['amountUSD'] = stats.amountUSD.apply(round)
stats.sort_values('amountUSD', ascending=False).loc[stats.amountUSD>1000]
```

```python
stats2 = stats.reset_index()
accts = set(list(stats2['from']) + list(stats2.to))
len(accts)
```

```python
long_list = []
for i, address in enumerate(accts):

  url = f"https://api.polkaholic.io"
  next_page = f"/account/xcmtransfers/{address}"

  r = requests.get(url=url+next_page, headers=headers).json()
  print(f"{i} Address {address}: {len(r)}")
  long_list.extend(r)

xcm = pd.DataFrame(long_list).sort_values('destTS', ascending=False)
xcm['timestamp'] = pd.to_datetime(xcm.destTS, unit="s")
#interesting = ['timestamp',*xcm.columns]
xcm
```

Let's aggregate all the xtransfer by chainID and currency first, counting the from and to addresses, figuring out when it started (min timestamp) and of course, the sum of monies received.

```python
import json
xcm['timestamp'] = pd.to_datetime(xcm.destTS, unit="s")
xcm['currency'] = xcm.asset.apply(lambda t: json.loads(t).get('Token'))
interesting_time = "2022-08-13 22:00:00"
interesting_df = xcm.loc[xcm.timestamp > interesting_time]

stats = interesting_df.groupby(['chainID', 'currency']).agg({
    'amountReceivedUSD' : 'sum',
    'amountReceived' : 'sum',
    'timestamp' : 'min',
    'fromAddress': 'nunique',
    'destAddress': 'nunique',
})
stats['amountUSD'] = stats.amountReceivedUSD.apply(round)
stats['amountReceived'] = stats.amountReceived.apply(round, ndigits=2)
stats.sort_values('amountReceived', ascending=False)
```

```python
import json
xcm['timestamp'] = pd.to_datetime(xcm.destTS, unit="s")
xcm['currency'] = xcm.asset.apply(lambda t: json.loads(t).get('Token'))
interesting_time = "2022-08-13 22:00:00"
interesting_df = xcm.loc[xcm.timestamp > interesting_time]
stats = interesting_df.groupby(['chainID', 'fromAddress', 'destAddress', 'currency']).agg({
    'amountReceivedUSD' : 'sum',
    'amountReceived' : 'sum',
    'timestamp' : 'min'
})
stats['amountUSD'] = stats.amountReceivedUSD.apply(round)
stats['amountReceived'] = stats.amountReceived.apply(round, ndigits=2)
stats.sort_values('amountReceived', ascending=False)
```
